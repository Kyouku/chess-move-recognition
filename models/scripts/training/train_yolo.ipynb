```python
from pathlib import Path
from ultralytics import YOLO

# Paths
YOLO_OUTPUT_DIR = Path("./yolo11_runs")
YOLO_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

YOLO_DATA_YAML = "./Automated-Chess-Move-Recognition-yolo11/data.yaml"

model = YOLO("yolo11s.pt")

results = model.train(
    data=YOLO_DATA_YAML,  # dataset config yaml
    project=str(YOLO_OUTPUT_DIR),  # root folder for detections and weights
    name="yolo11_chess",  # experiment name (subfolder)

    epochs=160,
    imgsz=640,  # roughly matches your RFDETR resolution=640
    batch=16,
    workers=12,

    patience=20,  # early stopping patience (like your 20)
    lr0=1e-3,  # starting lr (YOLO default is higher than RFDETR)
    weight_decay=1e-4,
    device="0",  # first CUDA GPU

    optimizer="auto",
)
```
